```python
def download_transparency_report(isin, year):
    
    '''
    The function uses selenium to download a dynamic table from ESMA's transparency report.

    Input
    isin: isin as string.
    year: yyyy as string.

    Output
    Returns one row dataframe.
    '''

    import pip

    try:
        __import__('selenium')
    except ImportError:
        pip.main(['install', 'selenium'])

    try:
        __import__('pandas')
    except ImportError:
        pip.main(['install', 'pandas'])

    from selenium import webdriver
    from selenium.webdriver.common.by import By
    import time
    import pandas as pd
    
    url = f"https://registers.esma.europa.eu/publication/details?core=esma_registers_fitrs_equities&docId={isin}YEAR{year}0101000000{year}1231000000ESMA"

    # Set up WebDriver #(make sure to have the appropriate webdriver installed)
    driver = webdriver.Chrome() 

    # Open URL
    driver.get(url)

    time.sleep(3.5)

    # Find 'detailsParent'
    table = driver.find_element(By.ID, 'detailsParent')

    # Extract data
    rows = table.find_elements(By.TAG_NAME, 'tr')

    # Initialize dictionary
    data_dict = {}

    # Iterate through rows
    for row in rows:
        columns = row.find_elements(By.TAG_NAME, 'td')
        if len(columns) == 2:
            # Add key-value pair to the dictionary
            data_dict[columns[0].text] = columns[1].text

    # Convert to DataFrame
    df = pd.DataFrame([data_dict])

    # Close browser
    driver.quit()

    return df
```

```python
download_transparency_report(isin = 'DE0005140008', year = '2018')
```
